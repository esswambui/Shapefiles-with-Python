```python
import pandas as pd
import matplotlib.pyplot as plt #if using matplotlib
import plotly.express as px #if using plotly
import geopandas as gpd
```

```python
# Read the shape files.
ke_poly = gpd.read_file("/content/ken_admbnda_adm2_iebc_20191031.shp")
ke_lines = gpd.read_file("/content/ken_admbndl_admALL_iebc_itos_20191031.shp")
ke_points = gpd.read_file("/content/ken_admbndp_admALL_iebc_itos_20191031.shp")
```

```python
ke_poly.head()
```

```python
ke_points.head()
```

```python
# basic plot of polygons
ke_poly.plot(column="ADM2_EN", cmap="gist_earth")
```

```python
# Using subplots yo plot points and lines.
fig, ax = plt.subplots(figsize=(8,12))
ke_lines.plot(ax=ax)
ke_points.plot(ax=ax)

ax.axis('off')
```
